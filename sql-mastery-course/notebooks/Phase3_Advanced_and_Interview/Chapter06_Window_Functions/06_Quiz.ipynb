**Q1.** Explain in your own words: Window functions compute across related rows **without collapsing** them: `FUNC() OVER (PARTITION BY … ORDER BY …)`.

**Q2.** Explain in your own words: `ROW_NUMBER` (unique), `RANK` (gaps), `DENSE_RANK` (no gaps) for ranking.

**Q3.** Explain in your own words: `PARTITION BY` restarts the calculation per group — the key to per-group analytics.

**Q4.** Explain in your own words: **Top-N-per-group** = rank in a subquery/CTE, filter the rank (or Snowflake `QUALIFY`).

**Q5.** Explain in your own words: Aggregate windows compare each row to its group average/total; add `ORDER BY` for running totals.

**Q6.** Explain in your own words: `LAG`/`LEAD` compare to previous/next rows (period-over-period).
