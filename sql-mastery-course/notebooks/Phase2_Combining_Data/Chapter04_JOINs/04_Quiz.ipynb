**Q1.** Explain in your own words: Tables are split to avoid repetition; **JOINs recombine them** via key columns.

**Q2.** Explain in your own words: `INNER JOIN` keeps only matches; `LEFT JOIN` keeps all left rows (+NULLs for non-matches).

**Q3.** Explain in your own words: **Anti-join** = `LEFT JOIN … WHERE right.key IS NULL` → "rows with no match".

**Q4.** Explain in your own words: Chain joins for 3+ tables; use a **self join** for within-table relationships (employee↔manager).

**Q5.** Explain in your own words: **JOIN + GROUP BY** powers most analytics.

**Q6.** Explain in your own words: Qualify columns (`e.dept_id`), pick the right join type, and beware double-counting.
